- Melakukan handling missing value pada fitur Daily Time Spent on Site, Area Income, Daily Internet Usage dengan mengisinya menggunakan mean.

- Melakukan handling missing value pada fitur Male dengan menghapus data missing value.

## 2.2 Data Preprocessing

### 2.2.1 Drop Features

```python
# Menghapus Fitur yang Tidak Diperlukan
drop_features = ['Unnamed: 0']

# df.drop(drop_features, axis=1, inplace=True)
df_clean = df_clean.drop(drop_features, axis=1, inplace=False)
df_clean
```

### 2.2.2 Feature Encoding

```python
# Tampilkan kolom dengan tipe data object (kategorikal)
print("\nCategorical Features:", categorical_features)
```

```python
# Menggunakan get_dummies untuk melakukan one-hot encoding pada data categorical
df_clean = pd.get_dummies(df_clean, columns=['Male'], prefix=['Gender'])
df_clean = pd.get_dummies(df_clean, columns=categorical_features[1:])
```

```python
print(f'Result one-hote encoding:')
print(df_clean.columns)
```

### 2.2.3 Split Data Feature

```python
# Pilih kolom 'Clicked on Ad' sebagai target dan sisanya sebagai feature
y = df_clean['Clicked on Ad']
x = df_clean.drop('Clicked on Ad', axis=1)
```

### 2.2.4 Extraction Datetime

```python
# Mengubah tipe data pada fitur Timestamp menjadi datetime
x['Timestamp'] = pd.to_datetime(x['Timestamp'])
```

```python
# Membuat kolom baru berdasarkan data waktu
x['Year'] = x['Timestamp'].dt.year
x['Month'] = x['Timestamp'].dt.month
x['Week'] = x['Timestamp'].dt.isocalendar().week
x['Day'] = x['Timestamp'].dt.day
```

```python
# Tampilkan data setelah ekstraksi waktu
print("\nData after datetime extraction:")
x[['Timestamp', 'Year', 'Month', 'Week', 'Day']].sample(7)
```

```python
# Drop kolom Timestamp jika tidak diperlukan lagi
x = x.drop('Timestamp', axis=1)
```

# 3. Modeling

## 3.1 Split Data Train & Test

```python
# Split the data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# # Tampilkan hasil split
print("\nData Splitting:")
print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## 3.2 Experiment Machine Learning

### 3.2.1 Experiment 1 (Without Normalization)

```python
# Logistic Regression (tanpa normalisasi)
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train, y_train)
y_pred_log_reg = log_reg.predict(x_test)
```

```python
# Decision Tree (tanpa normalisasi)
tree_clf = DecisionTreeClassifier(random_state=42)
tree_clf.fit(x_train, y_train)
y_pred_tree = tree_clf.predict(x_test)
```

```python
# Random Forest (tanpa normalisasi)
rf_clf = RandomForestClassifier(random_state=42)
rf_clf.fit(x_train, y_train)
y_pred_rf = rf_clf.predict(x_test)
```

### 3.2.2 Experiment 2 (With Normalization)

```python
# Normalisasi data menggunakan Min-Max Scaler
scaler = MinMaxScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Logistic Regression (dengan normalisasi)
log_reg_scaled = LogisticRegression(max_iter=1000)
log_reg_scaled.fit(x_train_scaled, y_train)
y_pred_log_reg_scaled = log_reg_scaled.predict(x_test_scaled)
```

```python
# Decision Tree (dengan normalisasi)
tree_clf_scaled = DecisionTreeClassifier(random_state=42)
tree_clf_scaled.fit(x_train_scaled, y_train)
y_pred_tree_scaled = tree_clf_scaled.predict(x_test_scaled)
```

```python
# Random Forest (dengan normalisasi)
rf_clf_scaled = RandomForestClassifier(random_state=42)
rf_clf_scaled.fit(x_train_scaled, y_train)
y_pred_rf_scaled = rf_clf_scaled.predict(x_test_scaled)
```

## 3.3 Evaluation Model

```python
# Function untuk menampilkan confusion matrix
def plot_confusion_matrix(y_true, y_pred, title):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(title)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.show()
```

### 3.3.1 Evaluation Experiment 1

```python
# Evaluasi logistic regression (tanpa normalisasi)
plot_confusion_matrix(y_test, y_pred_log_reg, "Logistic Regression (Without Normalization)")
```

```python
# Evaluasi decision tree (tanpa normalisasi)
plot_confusion_matrix(y_test, y_pred_tree, "Decision Tree (Without Normalization)")
```

```python
# Evaluasi random forest (tanpa normalisasi)
plot_confusion_matrix(y_test, y_pred_rf, "Random Forest (Without Normalization)")
```

### 3.3.2 Evaluation Experiment 2

```python
# Evaluasi logistic regression (dengan normalisasi)
plot_confusion_matrix(y_test, y_pred_log_reg_scaled, "Logistic Regression (With Normalization)")
```

```python
# Evaluasi decision tree (dengan normalisasi)
plot_confusion_matrix(y_test, y_pred_tree_scaled, "Decision Tree (With Normalization)")
```

```python
# Evaluasi random forest (dengan normalisasi)
plot_confusion_matrix(y_test, y_pred_rf_scaled, "Random Forest (With Normalization)")
```

### 3.3.3 Accuracy

```python
models = {
    "Logistic Regression (No Normalization)": y_pred_log_reg,
    "Decision Tree (No Normalization)": y_pred_tree,
    "Random Forest (No Normalization)": y_pred_rf,
    "Logistic Regression (With Normalization)": y_pred_log_reg_scaled,
    "Decision Tree (With Normalization)": y_pred_tree_scaled,
    "Random Forest (With Normalization)": y_pred_rf_scaled
}

for name, y_pred in models.items():
    print(f"\n=== {name} ===")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

- Akurasi tertinggi didapatkan oleh model Random Forest, baik dengan normalisasi maupun tanpa normalisasi. Dimana akurasinya yakni sebesar 95.5%.

- Model Random Forest ini menunjukkan performa yang sangat baik dengan akurasi tinggi. Ini berarti model jarang melakukan kesalahan baik dalam mendeteksi klik maupun tidak klik.

- Dengan precision 96.1% dan recall 95.1%, model Random Forest ini memiliki keseimbangan yang baik. Artinya, ketika model memprediksi pengguna akan mengklik, prediksi tersebut sangat andal (tinggi precision). Pada saat yang sama, model juga berhasil menangkap sebagian besar dari semua klik yang benar-benar terjadi (tinggi recall).

## 4.2 Business Simulation

```python
# Data untuk simulasi
jumlah_user = 10000
biaya_per_user = 10                 # Biaya marketing per user dalam USD
conversion_rate_tanpa_ml = 0.05     # 5% conversion tanpa model ML
harga_produk = 100                  # Harga produk dalam USD
```

### 4.2.1 Simulation Without Model

```python
# Simulasi Bisnis Tanpa Model Machine Learning
cost_tanpa_ml = jumlah_user * biaya_per_user
revenue_tanpa_ml = jumlah_user * conversion_rate_tanpa_ml * harga_produk
profit_tanpa_ml = revenue_tanpa_ml - cost_tanpa_ml
```

```python
# Menampilkan hasil simulasi
print("\nSimulasi Tanpa Model Machine Learning")
print(f"Cost\t: ${cost_tanpa_ml}")
print(f"Revenue\t: ${revenue_tanpa_ml}")
print(f"Profit\t: ${profit_tanpa_ml}")
```

### 4.2.2 Simulation With Model

```python
# Melakukan prediksi dengan menggunakan model terbaik
y_pred_rf_scaled = rf_clf_scaled.predict(x_test_scaled)

# Confusion matrix akan memberikan nilai TN, FP, FN, dan TP
tn, fp, fn, tp = confusion_matrix(y_test, y_pred_rf_scaled).ravel()
```

```python
# Menghitung nilai conversion rate
conversion_rate_dengan_ml = tp / (tp + fp + fn + tn)
conversion_rate_dengan_ml
```

```python
# Simulasi Bisnis Dengan Model Machine Learning
cost_dengan_ml = jumlah_user * biaya_per_user
revenue_dengan_ml = jumlah_user * conversion_rate_dengan_ml * harga_produk
profit_dengan_ml = revenue_dengan_ml - cost_dengan_ml
```

```python
# Menampilkan hasil simulasi
print("\nSimulasi Dengan Model Machine Learning")
print(f"Cost\t: ${cost_dengan_ml}")
print(f"Revenue\t: ${revenue_dengan_ml}")
print(f"Profit\t: ${profit_dengan_ml}")
```
